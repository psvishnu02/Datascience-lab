```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [2, 3, 3, 4, 5, 5, 5, 6, 6, 7, 7, 7, 7, 8, 9]

# Set a larger figure size
plt.figure(figsize=(8, 6))

# Plot histogram
plt.subplot(2, 2, 1)
plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Plot pie chart
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 40, 25, 10]
plt.subplot(2, 2, 4)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')

# Plot line graph
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.subplot(2, 2, 3)
plt.plot(x, y, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Graph')

# Generate data for the scatter plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Plot scatter plot
plt.subplot(2, 2, 2)
plt.scatter(x_scatter, y_scatter, color='red', marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import style

# Sample data
data = [2, 3, 3, 4, 5, 5, 5, 6, 6, 7, 7, 7, 7, 8, 9]

# Set a larger figure size
plt.figure(figsize=(8, 6))

# Plot histogram
plt.subplot(2, 2, 1)
plt.hist(data, bins=5, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')

# Plot pie chart
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 40, 25, 10]
plt.subplot(2, 2, 4)
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart')

# Plot line graph
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.subplot(2, 2, 3)
plt.plot(x, y, color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Line Graph')

# Generate data for the scatter plot
x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

# Plot scatter plot
plt.subplot(2, 2, 2)
plt.scatter( y_scatter, x_scatter, color='yellow', marker='o')
plt.scatter(x_scatter, y_scatter, color='pink', marker='o')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')

plt.suptitle(' Graphs in a single plot', fontsize=20)

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()

# x axis values 
x = [1,2,3,4,5,6] 
# corresponding y axis values 
y = [2,4,1,5,2,6] 

plt.subplot(2, 2, 1)
# plotting the points 
plt.plot(x, y, color='blue', linestyle='dashed', linewidth = 1, 
		marker='o', markerfacecolor='red', markersize=10) 

# setting x and y axis range 
plt.ylim(1,8) 
plt.xlim(1,8) 

# naming the x axis 
plt.xlabel('x - axis') 
# naming the y axis 
plt.ylabel('y - axis') 

# giving a title to my graph 
plt.title('Customized graph') 
style.use('ggplot')
# function to show the plot 
plt.show()
```
